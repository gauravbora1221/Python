# Other 2D Graphs

```python
n=np.array([0,1,2,3,4,5])
fig, axes=plt.subplots(1,4,figsize=(10,3.5))
axes[0].set_title('Scatter')
axes[0].scatter(x,x+0.25*np.random.randn(len(x)))

axes[1].set_title('Step')
axes[1].step(n,n**2,lw=2)

axes[2].set_title('Bar')
axes[2].bar(n,n**2,align='center',alpha=0.5)

axes[3].set_title('fill between')
axes[3].fill_between(x,x**2,x**3,color='g',alpha=0.5)
```

# Draw a Radar Chart

```python
fig=plt.figure(figsize=(5,5))
ax=fig.add_axes([0,0,0.6,0.6],polar=True)
t=np.linspace(0,2*np.pi,100)
ax.plot(t,t,color='green',lw=3)
```

```python
# HISTOGRAM

n=np.random.randn(10000)
fig,axes=plt.subplots(1,2,figsize=(8,4))
axes[0].set_title('default Histogram')
axes[0].hist(n)

axes[1].set_title('Cumulative Histogram')
axes[1].hist(n,cumulative=True,bins=50)
```

# DRAW CONTOR IMAGE

```python
import matplotlib
import matplotlib.cm as cm
import numpy as np
import matplotlib.pyplot as plt

delta=0.025
x=np.arange(-3.0,3.0,delta)
y=np.arange(-2.0,2.0,delta)
X,Y=np.meshgrid(x,y)
Z1=np.exp(-X**2 - Y**2)
Z2=np.exp(-(X-1)**2 -(Y-1)**2)
Z=(Z1-Z2)*2
print(X)
print(Y)
```

```python
fig,axes=plt.subplots()
CS=axes.contour(X,Y,Z)
axes.clabel(CS,inline=1,fontsize=8)
axes.set_title('contour')
```
