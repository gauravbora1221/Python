```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
c1=[2600,3000,3200,3600,4000,4500,5000,5500,6000]
c2=[55000,56500,61000,68000,72500,76400,80000,84000,87000]
```

```python
data={'area':c1,'price':c2}
df=pd.DataFrame(data)
```

```python
print(df[['area']])
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price)
```

```python
reg=linear_model.LinearRegression()
reg.fit(df.drop('price',axis='columns'),df.price)
```

```python
reg.predict([[3000]])
```

```python
y=reg.predict(df.drop('price',axis='columns'))
```

```python
plt.plot(df.area,y)
plt.scatter(df.area,df.price)
```
