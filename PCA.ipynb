```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.array([[2,1],[3,5],[4,3],[5,6],[6,7],[7,8]])
mean = np.mean(data,axis = 0)
feature = (data - mean)
covmatrix = np.cov(feature.T)
```

```python
eigen_values,eigen_vectors = np.linalg.eig(covmatrix)
principal_component = eigen_values[np.argmax(eigen_values)]
principal_component
```

```python
plt.scatter(data[:,0], data[:,1], alpha=0.5)
plt.quiver(mean[0], mean[1], eigen_vectors[0,0], eigen_vectors[1,0], scale=5, color='red')
plt.quiver(mean[0], mean[1], eigen_vectors[0,1], eigen_vectors[1,1], scale=5, color='blue')
```
